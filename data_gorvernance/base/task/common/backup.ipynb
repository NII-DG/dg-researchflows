# バックアップを設定する

所属機関の情報セキュリティポリシーや研究助成機関のデータポリシーなどを参照し、適切なタイミング、場所にデータのバックアップを設定するタスクです。

## バックアップの場所

### バックアップの場所を選択する

1. 事前に所属機関の情報セキュリティポリシーや研究助成機関のデータポリシーなどで、データの保存に関する規定を確認してください。
1. ストレージに求める条件（利用料金・容量・保存期間・バージョン管理の有無・対応フォーマット・共有機能等）を考慮してバックアップ先のストレージを選択してください。
1. 所属機関のネットワークドライブや、組織として契約したオンラインストレージが条件を満たしていれば優先的に使用してください。
1. 外部の商用クラウドサービス等を利用する場合は、情報セキュリティに関して条件設定の余地があるものを選び、機関内の外部委託に関する規程やクラウドサービスの利用に関する規程に反していないかあらかじめ確認してください。


### バックアップ場所の種類を検討する

研究データの保存場所の種類とそれぞれの特徴については表のとおりです。

| 保存場所 | 推奨度 | データの保存先として適切か |
| :---- | :----: | :---- |
| ・PC本体のHDD<br>・外部記憶デバイス<br>（USBメモリ・外付けHDD） | × | 破損・障害によるデータ消失のリスク、<br>紛失・盗難などのセキュリティリスクが高い |
| ・所属機関の提供する<br>ネットワークドライブ  | △ | 所属機関のシステム管理部門が提供し、<br>安定的に運用されているものであれば保存先として最適 |
| ・外部の提供する<br>オンラインストレージ  | 〇 | 組織が契約している場合など、情報セキュリティに関する<br>十分な条件設定の余地があれば適切<br>個人で登録して利用するものには注意が必要 |

バックアップ先は必ずしもひとつである必要はありません。
研究の段階に応じてバックアップ先を選択してください。

## バックアップ設定を行う

### バックアップの重要性を理解する

1. 不慮のトラブルで研究データを失った際にも、トラブルの影響を最小限に抑えることができる
1. 不用意なファイルの削除や上書き、システム障害にも対応できる


### バックアップの設定を行う

1. 既に利用しているストレージに付随するバックアップ機能がないか確認します。
1. バックアップは自動か手動か、OSの標準機能か専用のソフトウェアが必要か等を決めてください。ストレージの容量が不足していれば新たに用意してください。
1. 指定した範囲・頻度・保存先でバックアップがとれるよう設定してください。
1. バックアップを手動で実行し、バックアップにかかる時間、容量、保存先などが想定どおりか確認し、作成したバックアップから復元できるかどうかも確認してください。
1. バックアップの対象となるデータの容量、バックアップにかかる時間等を考慮して頻度を調節してください。（毎日差分バックアップで毎月1回フルバックアップをする、など）



### バックアップをするうえで注意すること

* 重要なデータの場合には災害対策として、物理的に離れた複数の場所にバックアップを取っていてください。
* バックアップ先のストレージを定期的に確認し、障害が発生していないか、データ容量を超過していないか確認してください。
* バックアップの実行ログを取得できるようにし、定期的にバックアップが実行されていることを確認してください。
